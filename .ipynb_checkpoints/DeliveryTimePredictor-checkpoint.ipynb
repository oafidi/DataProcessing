```python
df = pd.read_csv('CleanedData.csv')
df.head()
```

```python
# Calcul de la matrice de corrÃ©lation
df['Delivery_person_Ratings_complement'] = 6 - df['Delivery_person_Ratings']
quantitative_cols = [
    'Delivery_person_Age', 'Delivery_person_Ratings_complement',
    'Vehicle_condition', 'multiple_deliveries',
    'Time_taken(min)', 'Pickup_delay', 'Distance(Km)'
]
quant_df = df[quantitative_cols]
corr_matrix = quant_df.corr()
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
cmap = sns.diverging_palette(240, 10, as_cmap=True)
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap=cmap,
    center=0,
    linewidths=0.75,
    cbar_kws={"shrink": 0.8, "label": "CorrÃ©lation"},
    square=True
)
plt.title("CorrÃ©lation entre variables quantitatives", fontsize=16, weight='bold')
plt.xticks(rotation=25, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
correlations = df[quantitative_cols].corr()['Time_taken(min)'].drop('Time_taken(min)')
selected_quantitative_features = correlations[abs(correlations) >= 0.29].index.tolist()
print("Variables Ã  standardiser :", selected_quantitative_features)
```

```python

quantitative_cols = [
    'Delivery_person_Age', 'Delivery_person_Ratings',
    'Vehicle_condition', 'multiple_deliveries',
    'Time_taken(min)', 'Pickup_delay', 'Distance(Km)'
]

x_dropdown = widgets.Dropdown(options=quantitative_cols, description='X:', value='Distance(Km)')
y_dropdown = widgets.Dropdown(options=quantitative_cols, description='Y:', value='Time_taken(min)')

def update_plot(x_col, y_col):
    if len(df) >= 100:
        data = df.sample(100, random_state=42)
    else:
        data = df.copy()
    
    fig = px.scatter(
        data, x=x_col, y=y_col,
        title=f"{y_col} vs {x_col}",
        labels={x_col: x_col, y_col: y_col},
        color='multiple_deliveries',
        color_continuous_scale='Viridis',
        height=500
    )
    fig.update_traces(marker=dict(size=10, opacity=0.7))
    fig.show()

ui = widgets.HBox([x_dropdown, y_dropdown])
out = widgets.interactive_output(update_plot, {'x_col': x_dropdown, 'y_col': y_dropdown})

display(ui, out)
```

```python
def standard_scaler_from_scratch(df, columns):
    df_scaled = df.copy()
    for col in columns:
        mean = df[col].mean()
        std = df[col].std()
        df_scaled[col + '_scaled'] = (df[col] - mean) / std
        print(f"{col}: Moyenne = {mean:.2f}, Variance = {std:.2f}")
    return df_scaled
df_scaled = standard_scaler_from_scratch(df, selected_quantitative_features)
```

```python
cov_matrix = np.cov(df_scaled[selected_quantitative_features].T)
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)
sorted_idx = np.argsort(eigen_values)[::-1]
eigen_values = eigen_values[sorted_idx]
eigen_vectors = eigen_vectors[:, sorted_idx]
explained_variance_ratio = eigen_values / np.sum(eigen_values)
cumulative_variance = np.cumsum(explained_variance_ratio)
```

```python
explained_variance_ratio
```

```python
X = df[selected_quantitative_features].values
X_pca = X.dot(eigen_vectors[:, :2])
df[['PC1', 'PC2']] = pd.DataFrame(X_pca, columns=['PC1', 'PC2'], index=df.index)
df_sample = df.sample(n=500, random_state=42)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(explained_variance_ratio)+1), cumulative_variance*100, marker='o', linestyle='--', color='orange')
plt.title('Variance expliquÃ©e par composante')
plt.xlabel('Composante principale')
plt.ylabel('Pourcentage de variance')
plt.xticks([1, 2, 3, 4])
plt.grid(True)
plt.show()
```

```python
qualitative_vars = [
    'Delivery_person_ID',
    'Weather_conditions',
    'Road_traffic_density',
    'Type_of_order',
    'Type_of_vehicle',
    'Festival',
    'Area_Type',
    'Day_of_week',
    'Is_month_start',
    'Is_month_end',
    'Is_weekend'
]
df[['Delivery_person_Age_scaled', 'Delivery_person_Ratings_complement_scaled', 'multiple_deliveries_scaled', 'Distance(Km)_scaled']] = pd.DataFrame(df_scaled, columns=['Delivery_person_Age_scaled', 'Delivery_person_Ratings_complement_scaled', 'multiple_deliveries_scaled', 'Distance(Km)_scaled'], index=df.index)
```

```python
anova_results = []

for var in qualitative_vars:
    if df[var].nunique() > 1:
        groups = [g['Time_taken(min)'].dropna().values for _, g in df.groupby(var)]
        if len(groups) > 1:
            try:
                f_stat, p_val = f_oneway(*groups)
                anova_results.append({
                    'Variable': var,
                    'F-statistic': round(f_stat, 2),
                    'p-value': round(p_val, 5),
                    'Significatif': 'Oui' if p_val < 0.05 else 'Non'
                })
            except Exception as e:
                anova_results.append({
                    'Variable': var,
                    'F-statistic': None,
                    'p-value': None,
                    'Significatif': f'Erreur : {e}'
                })
anova_df = pd.DataFrame(anova_results).sort_values('p-value', na_position='last')
anova_df
```

```python
anova_df = anova_df[anova_df['Significatif'] == 'Oui']
selected_qualitative_features = list(anova_df['Variable'])

qualitative_vars = [
    'Weather_conditions',
    'Road_traffic_density',
    'Type_of_order',
    'Type_of_vehicle',
    'Area_Type',
    'Day_of_week',
]
cat_dropdown = widgets.Dropdown(
    options=qualitative_vars,
    description='CatÃ©gorie:',
    value='Road_traffic_density'
)

def update_bar_plot(cat_var):
    counts = df[cat_var].value_counts().reset_index()
    counts.columns = [cat_var, 'Nombre']

    fig = px.bar(
        counts,
        x=cat_var,
        y='Nombre',
        title=f"RÃ©partition par {cat_var}",
        text='Nombre',
        color=cat_var,
        color_discrete_sequence=px.colors.qualitative.Vivid,
        height=500
    )
    
    fig.update_traces(textposition='outside')
    fig.update_layout(
        xaxis_title=cat_var,
        yaxis_title='Nombre',
        showlegend=False
    )
    
    fig.show()

out_bar = widgets.interactive_output(update_bar_plot, {'cat_var': cat_dropdown})
display(widgets.HBox([cat_dropdown]), out_bar)
```

```python
def squared_error(X_train,y_train, w, b):
    predictions = np.dot(X_train.iloc[:, :-1], w) + b
    errors = predictions - y_train
    return np.sum(errors**2) / (2 * len(X_train))
```

```python
def gradient_descent(X_train,y_train, w, b, alpha):
    
    predictions = np.dot(X_train.iloc[:, :-1], w) + b
    errors = predictions - y_train
    gradient_w = np.dot(X_train.iloc[:, :-1].T, errors) / len(X_train)
    gradient_b = np.sum(errors) / len(X_train)
    w -= alpha * gradient_w
    b -= alpha * gradient_b
    
    return w, b
```

```python
w = np.zeros(X_train.shape[1] - 1)
b = 0 

alpha = 0.01  
iterations = 4000 
cost_track = []
iteration_track = []
# EntraÃ®nement du modÃ¨le
for iteration in range(iterations):
    w, b = gradient_descent(X_train,y_train, w, b, alpha)
    if iteration % 100 == 0: 
        cost = squared_error(X_train,y_train, w, b)
        cost_track.append(cost)
        iteration_track.append(iteration)
        print(f'L\'iteration {iteration}, CoÃ»t: {cost}')
```

```python
plt.figure()
plt.plot(iteration_track, cost_track)
plt.xlabel("iteration")
plt.ylabel("Cost")
```

```python
print(f'Poids optimaux: {w}')
print(f'Biais optimal: {b}')
```

```python
class KNNRegressor:
    def __init__(self, k=3):
        self.k = k
        
    def fit(self, X_train, y_train):
        self.X_train = np.array(X_train)
        self.y_train = np.array(y_train)
        
    def predict(self, X_test):
        X_test = np.array(X_test)
        predictions = []
        
        for x in X_test:
            distances = np.sqrt(np.sum((self.X_train - x)**2, axis=1))
            k_indices = np.argsort(distances)[:self.k]
            k_nearest_targets = self.y_train[k_indices]
            prediction = np.mean(k_nearest_targets)
            predictions.append(prediction)
        
        return np.array(predictions)
```

```python
k = int(np.sqrt(len(X_train)))
model = KNNRegressor(k=k)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
for i in range(20):
    print(f"Y_test: {y_test[i]}   Y_predicted: {y_pred[i]}")
```

```python
efficacite = r2_score(y_test, y_pred)
print(f"Taux d'efficacitÃ© (RÂ²) : {efficacite * 100:.2f}%")
```
